```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame

df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data

df_market_data.head(10)
```

```python
# Generate summary statistics

df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame

df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file

market_data_scaled = StandardScaler().fit_transform(df_market_data[["price_change_percentage_24h", "price_change_percentage_7d", "price_change_percentage_14d", "price_change_percentage_30d", "price_change_percentage_60d", "price_change_percentage_200d", "price_change_percentage_1y"]])
```

```python
# Create a DataFrame with the scaled data

df_market_data_scaled = pd.DataFrame(
    market_data_scaled, 
    columns=["price_change_percentage_24h", "price_change_percentage_7d", "price_change_percentage_14d", "price_change_percentage_30d", "price_change_percentage_60d", "price_change_percentage_200d", "price_change_percentage_1y"]
)

# Copy the crypto names from the original data

df_market_data_scaled["coin_id"] = df_market_data.index

# Set the coinid column as index

df_market_data_scaled = df_market_data_scaled.set_index("coin_id")

# Display sample data

df_market_data_scaled.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11

k_values = list(range(1, 11))

print(k_values)
```

```python
# Create an empty list to store the inertia values

inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list

for i in k_values:
    k_model = KMeans(n_clusters=i, random_state=0)
    k_model.fit(df_market_data_scaled)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve

elbow_curve = {"k": k_values, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve

elbow_curve_df = pd.DataFrame(elbow_curve)

elbow_curve_df.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.

elbow_curve_df.hvplot.line(
    x = "k",
    y = "inertia",
    title = "Elbow Curve Using Original Data",
    xticks = k_values
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The best value for 'k' is 4 clusters.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k

model_4 = KMeans(n_clusters=4, random_state = 0)
```

```python
# Fit the K-Means model using the scaled data

model_4.fit(df_market_data_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data

k_4 = model_4.predict(df_market_data_scaled)

# Print the resulting array of cluster values.

k_4
```

```python
# Create a copy of the DataFrame

df_market_data_scaled_predictions_4 = df_market_data_scaled.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters

df_market_data_scaled_predictions_4["segments"] = k_4

# Display sample data

df_market_data_scaled_predictions_4.head(10)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

df_market_data_scaled_predictions_4.hvplot.scatter(
    x = "price_change_percentage_24h",
    y = "price_change_percentage_7d",
    by = "segments",
    title = "Cryptocurrency Predictions (Original Data)",
    hover_cols = ["coin_id"]
)
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.

pca = PCA(n_components = 3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.

market_data_pca = pca.fit_transform(df_market_data_scaled)

# View the first five rows of the DataFrame. 

market_data_pca[0:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.

pca.explained_variance_ratio_
```

```python
#Calculate total explained variance of the three principal components

0.3719856 + 0.34700813 + 0.17603793
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance of the three principal components is 89.5% reducing the overall accuracy 10.5%.  (Note:  0.3719856 + 0.34700813 + 0.17603793 = .89503166 or 89.5%)

```python
# Create a new DataFrame with the PCA data.
# Creating a DataFrame with the PCA data

df_market_data_pca = pd.DataFrame(market_data_pca, columns=["PCA1", "PCA2", "PCA3"])
    
# Copy the crypto names from the original data

df_market_data_pca["coin_id"] = df_market_data.index

# Set the coinid column as index

df_market_data_pca = df_market_data_pca.set_index("coin_id")

# Display sample data

df_market_data_pca.head(10)
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11

k_values_pca = list(range(1, 11))

print(k_values_pca)
```

```python
# Create an empty list to store the inertia values

inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list

for i in k_values_pca:
    k_model = KMeans(n_clusters = i, random_state = 0)
    k_model.fit(df_market_data_pca)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve

elbow_data_pca = {"k_pca": k_values_pca, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve

df_elbow_pca = pd.DataFrame(elbow_data_pca)

df_elbow_pca.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.

elbow_pca_plot = df_elbow_pca.hvplot.line(
    x = "k_pca",
    y = "inertia",
    title = "Elbow Curve Using PCA Data",
    xticks = k_values_pca
)

elbow_pca_plot
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** The best value for 'k' using the PCA data is 4 clusters.

* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No, the number of clusters are the same among the two data sets.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k

model_v1 = KMeans(n_clusters=4)
```

```python
# Fit the K-Means model using the PCA data

model_v1.fit(df_market_data_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data

market_data_clusters_4 = model_v1.predict(df_market_data_pca)

# Print the resulting array of cluster values.

print(market_data_clusters_4)
```

```python
#### 4 Clusters ####
# Create a copy of the DataFrame with the PCA data

df_market_data_pca_predictions_4 = df_market_data_pca.copy()

# Add a new column to the DataFrame with the predicted clusters

df_market_data_pca_predictions_4["predicted_clusters"]= market_data_clusters_4

# Display sample data

df_market_data_pca_predictions_4.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

df_market_data_pca_predictions_4.hvplot.scatter(
    x = "PCA1",
    y = "PCA2",
    by = "predicted_clusters",
    title = "Scatter Plot by Coin ID Segment",
    hover_cols = "coin_id"
)
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
# Composite plot to contrast the Elbow curves

# Create the first plot for the original data

elbow_plot_original = elbow_curve_df.hvplot.line(x="k", y="inertia", title="Elbow Curve Using Original Data")

# Create the second plot for the PCA data

elbow_plot_pca = df_elbow_pca.hvplot.line(x="k_pca", y="inertia", title="Elbow Curve Using PCA Data")

# Combine the two plots into a composite plot

composite_elbow_plot = elbow_plot_original + elbow_plot_pca

# Display the composite plot

composite_elbow_plot
```

```python
# Composite plot to contrast the clusters

# Create scatter plots for original data and PCA data clusters
original_cluster_plot = df_market_data_scaled_predictions_4.hvplot.scatter(x="price_change_percentage_24h", y="price_change_percentage_7d", by="segments", title='Original Data Clusters')
pca_cluster_plot_4 = df_market_data_pca_predictions_4.hvplot.scatter(x='PCA1', y='PCA2', by='predicted_clusters', title='PCA Data Clusters')

# Combine the scatter plots into a composite plot
composite_cluster_plot_v1 = original_cluster_plot + pca_cluster_plot_4

# Display the composite plot
composite_cluster_plot_v1
```
